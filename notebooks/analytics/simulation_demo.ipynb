```python
from TerraFin import configure, load_terrafin_config
from TerraFin.analytics.simulation.stochastic.gbm import run_base_gbm
from TerraFin.data import DataFactory
from TerraFin.data.utils.filters import date_filter
from TerraFin.interface.prediction_plots import timeseries_with_predictions


configure()
config = load_terrafin_config()

test_data = DataFactory().get_market_data("Nasdaq")
test_data = date_filter(test_data, "2023-01-01", "2025-6-30")

sim_results = run_base_gbm(test_data)

fig = timeseries_with_predictions(test_data, sim_results)
fig.show()
```
